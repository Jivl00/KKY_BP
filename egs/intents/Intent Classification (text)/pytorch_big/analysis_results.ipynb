```python
import numpy as np
# heatmap of neural network architecture and number of classes and accuracy
mapping = {}
with open('analysis.txt', 'r', encoding='utf-8') as f:
    # matrix 5x5
    accuracy = []
    for line in f:
        vals = line.split('\t')
        accuracy.append(float(vals[2]))


accuracy = np.array(accuracy).reshape(5, 5)
print(accuracy)
# reverse columns
accuracy = accuracy[:, ::-1]
print(accuracy)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=2, rc={"lines.linewidth": 5.5})

fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(np.array(accuracy*100).reshape(5, 5), annot=True, fmt='.2f', cmap='Blues', cbar=False)
# set axis labels
ax.set_xticklabels(['30', '60', '90', '120', '150'])
ax.set_yticklabels(['[512,256]', '[256,128]', '[128,64]', '[64,32]', '[32,16]'])
ax.set_xlabel('Number of classes')
ax.set_ylabel('Neural network architecture')
ax.set_title('Accuracy of neural network in %')
plt.savefig('heatmap.pdf', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
# heatmap of neural network accuracy depending on number of classes and number of samples per class
mapping = {}
with open('analysis2.txt', 'r', encoding='utf-8') as f:
    # matrix 5x5
    accuracy = []
    for line in f:
        vals = line.split('\t')
        accuracy.append(float(vals[2]))


accuracy = np.array(accuracy).reshape(5, 5)
print(accuracy)
accuracy = accuracy[::-1, :]
accuracy = accuracy[:, ::-1]
print(accuracy)
```
